```python
import pandas as pd
import os

## Listing Directory Files
testlist = os.listdir("C:/Users/Malcolm/OneDrive - University of Cambridge/Desktop/## Cambridge Academics/NST IB/Computational Things/20220620 Wright Project/CHEM-EDB Data Files")

## Generating Headers
headers = pd.read_csv("C:/Users/Malcolm/OneDrive - University of Cambridge/Desktop/## Cambridge Academics/NST IB/Computational Things/20220620 Wright Project/CHEM-EDB Data Files/pubchem.chembl.dataset4publication_inchi_smiles_0.csv",
                      sep = '\t')
headers = headers.columns [[1, 4, 8, 11]]
headers = headers.tolist()

##********** Iterating Across Directory Files (Run ONCE for filtration of specific protein)
def extract_gene_file(gene_name):
    gene_compiled_list = []
    for index, file in enumerate(testlist):
        bigfile = pd.read_csv("C:/Users/Malcolm/OneDrive - University of Cambridge/Desktop/## Cambridge Academics/NST IB/Computational Things/20220620 Wright Project/CHEM-EDB Data Files/" + file, sep = '\t',
                                  header = None)
        print(index)
        # Trying to identify protein targets and extract relevant rows
        gene_indexes = []
        for idx, gene in enumerate(bigfile.iloc[:,8]):
            if gene_name.upper() in gene:
                gene_indexes.append(idx)
        gene_list = bigfile.iloc[gene_indexes, [1, 4, 8, (len(bigfile.columns)-1)]]

        # Breaking down df into list and appending it as recommended
        for idx, row in gene_list.iterrows():
            gene_compiled_list.append(row)
        
    ## Final Printing.
    df_final = pd.DataFrame(gene_compiled_list)
    df_final.columns = headers
    df_final.to_csv(gene_name + "_file.csv", index = False)

extract_gene_file('egfr')
extract_gene_file("brca")
```

```python
import csv
from rdkit import Chem

## Extracting SMILES and pIC50 values from the different csv target files.
targets = ["egfr", 'brca']
data = dict([(p, list(csv.reader(open("C:/Users/Malcolm/OneDrive - University of Cambridge/Desktop/## Cambridge Academics/NST IB/Computational Things/20220620 Wright Project/%s_file.csv" % p)))[1:]) for p in targets])
molecules = dict([(p, [Chem.MolFromSmiles(d[3]) for d in data[p]]) for p in targets])
chembl_id = dict([(p, [d[0] for d in data[p]]) for p in targets])
pic50 = dict([(p, [d[1] for d in data[p]]) for p in targets])
```

```python
import numpy as np
from itertools import chain
from rdkit import DataStructs
from rdkit.Chem.Fingerprints import FingerprintMols


totalmolecules_clean = chain.from_iterable([x for x in molecules[p] if x is not None] for p in targets)
fps = [Chem.RDKFingerprint(x) for x in totalmolecules_clean]
# Make sure to remove Nones.

## Calculating Similarities
similarities = np.array([[DataStructs.FingerprintSimilarity(fps[i], fps[j]) for i in range(1,len(fps))] for j in range(1,len(fps))])

```

```python
from sklearn.manifold import TSNE

## Performing tSNE
embedded = TSNE(n_components = 2).fit_transform(similarities)
```

```python
from matplotlib import pyplot as plt

## Graphing
fig, ax = plt.subplots(figsize = (10, 10))
ax.scatter(embedded[0:len(similarities), 0], embedded[0:len(similarities), 1],
           marker = '.', linewidth = 0.01, color = 'r')
ax.set_title('2D Fingerprint of EGFR Inhibitors', color = 'red', font = 'Times New Roman', fontsize = 16)
ax.legend()

plt.show()
```
